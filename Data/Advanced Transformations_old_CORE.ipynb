# 1. Compare the average weight of super heros who have Super Speed to those who do not.

```python
ohe_super_powers_df['Super Speed']
```

```python
super_speed_df = ohe_super_powers_df[ohe_super_powers_df['Super Speed'] == True]
```

```python
super_speed_df
```

```python
super_speed_df = super_speed_df.rename(columns={"hero_names":"Hero"})
```

```python
super_speed_df
```

```python
super_info_df.head()
```

```python
HERO_W_SS_MERGED = super_speed_df.merge(super_info_df, how='inner', on=["Hero"])
HERO_W_SS_MERGED
```

```python
#Finding those Super Heros who don't have Super Speed.

# filtered ohe_super_speed_df to those without Super Speed and saved to new df
no_super_speed_df = ohe_super_powers_df[ohe_super_powers_df['Super Speed'] == False]

#change the name of the column hero_name to Hero so dataframes can be merged and save\replaced same df
no_super_speed_df = no_super_speed_df.rename(columns={"hero_names":"Hero"})

#Inner Merged no_super_speed_df to super_info_df on Hero column and saved to new df
HERO_NO_SS_MERGED = no_super_speed_df.merge(super_info_df, how='inner', on=["Hero"])
```

```python
print(f"Average weight of Super Heros with Super Speed is:  {HERO_W_SS_MERGED['Weight'].mean().round(3)}")
print(f"Average weight of Super Heros without Super Speed is:  {HERO_NO_SS_MERGED['Weight'].mean().round(3)}")
```
